```python
df = pd.read_csv(r"E:\Data_Science _master\ML\prac_project\Concrete Compressive Strength Prediction\notebooks\data\concrete_data.csv")
```

```python
df.head(123)
```

```python
# Detect outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Filtering outliers
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"Original data shape: {df.shape}")
print(f"Data shape after outlier removal: {df_no_outliers.shape}")
```

```python
df__=df_no_outliers
```
