```python
import pandas as pd
```

```python
df= pd.read_csv(r"E:\Data_Science _master\ML\prac_project\Concrete Compressive Strength Prediction\notebooks\data\concrete_data.csv")
```

```python
df.head()
```

```python
# Detect outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Filtering outliers
df__ = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

```

```python
df__
```

```python
X = df__.drop("concrete_compressive_strength", axis=1)
```

```python
X
```

```python
y= df__["concrete_compressive_strength"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25)
```

```python
X_train
```

```python
X_train.head()
```

```python
X_test
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline= Pipeline(
    
   steps= [
      
       
        ("Scaler",StandardScaler())   
   ] 
    
)
```

```python
X_train_scaled = num_pipeline.fit_transform(X_train)
X_test_scaled = num_pipeline.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
models = {
    "DecisionTreeRegressor":DecisionTreeRegressor(),
    "SVR":SVR(),
    "RandomForestRegressor":RandomForestRegressor(),
    "LinearRegression":LinearRegression()
    
}


for name, model in models.items():
    model.fit(X_train_scaled,y_train)
    
    y_pred = model.predict(X_test_scaled)
    
    r2= r2_score(y_test,y_pred)
    
    print(f"{name} R² Score: {r2:.4f}")
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [10, 20, 30, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ['sqrt', 'log2', None],  # Use 'sqrt' or 'log2' instead of 'auto'
    "bootstrap": [True, False]
}
```

```python
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
model.fit(X_train_scaled,y_train)
```
